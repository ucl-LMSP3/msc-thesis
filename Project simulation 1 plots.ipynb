```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data1_mean = pd.read_csv('1.1.csv', index_col='Model')
data1_std = pd.read_csv('1.2.csv', index_col='Model')
data2_mean = pd.read_csv('2.1.csv', index_col='Model')
data2_std = pd.read_csv('2.2.csv', index_col='Model')
data3_mean = pd.read_csv('3.1.csv', index_col='Model')
data3_std = pd.read_csv('3.2.csv', index_col='Model')
data4_mean = pd.read_csv('4.1.csv', index_col='Model')
data4_std = pd.read_csv('4.2.csv', index_col='Model')
```

```python
plt.rcParams["font.size"] = "18"
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 20))
data1_mean.plot.bar(ax=ax1, yerr=data1_std, capsize=4, fontsize=18, legend=False, title='Experiment 1: Friedman3 function with shared GP')
data2_mean.plot.bar(ax=ax2, yerr=data2_std, capsize=4, fontsize=18, legend=False, title='Experiment 2: Friedman3 function with individual GP')
data3_mean.plot.bar(ax=ax3, yerr=data3_std, capsize=4, fontsize=18, legend=False, title='Experiment 3: Linear function with shared GP')
data4_mean.plot.bar(ax=ax4, yerr=data4_std, capsize=4, fontsize=18, legend=False, title='Experiment 4: Linear function with individual GP')
x_axis1 = ax1.axes.get_xaxis()
x_axis1.set_visible(False)
x_axis2 = ax2.axes.get_xaxis()
x_axis2.set_visible(False)
fig.legend([ax1, ax2, ax3, ax4],
           labels=['Mean RMSE (Interpolation)', 'Mean RMSE (Extrapolation)', 'Mean RMSE (F(x))'],
           bbox_to_anchor=(1.19, 0.95))
fig.tight_layout()
plt.show()
```
